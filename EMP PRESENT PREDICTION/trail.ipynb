```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
import pickle
from sklearn import preprocessing
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('absent.csv')
df.head()
```

```python
df.Day.unique()
```

```python
df1 = df[['Name','Day']]
df2 = df[['Year','Month','Date','Present']]
df1.shape, df2.shape
```

```python
encode = ['Name','Day']
for col in encode:
    dummy = pd.get_dummies(df1[col], prefix=col)
    df1 = pd.concat([df1,dummy], axis=1)
    del df1[col]
```

```python
df1.shape
```

```python
df1.head()
```

```python
df2.head()
```

```python
df_final = pd.concat([df1,df2], axis=1)
df_final.head()
```

# modeling started

```python
X=df_final.drop('Present', axis=1)
y=df_final['Present'].copy()

X.shape, y.shape
```

```python
lg=LogisticRegression()
lg.fit(X,y)
```

```python
print(X.columns)
```

```python
pickle.dump(lg, open('present_absent.pkl', 'wb'))
```
